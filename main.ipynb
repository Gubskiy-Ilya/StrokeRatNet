# StrokeRatNet

AI for automatic segmentation of the cerebral infarction core on the MCAO model using isotropic T2-weighted images. The percentage of involvement of the blood supply areas of the right ICA branches in the stroke is additionally assessed.

#### Scanning protocol requirements
T2-weighted images from ClinScan 7T MR-system (Bruker BioSpin), SPACE pulse sequence, TR\TE = 4000\251 ms; voxel size 0.2х0.2х0.2 mm.

### Step 1
Place DICOM data in '**data**' folder

```python
from under_the_hood import dcm_to_nii_n4, stroke_volume, ica_atlas, ica_blood_supply
import warnings
warnings.filterwarnings("ignore")
dcm_to_nii_n4('./data/')
```

### Step 2
Run segmentation, nii files and final image will appear in the '**results**' folder (image.nii, label.nii, segmentation.png)

```python
stroke_volume()
```

### Step 3
Run registration of results to template, nii file will appear in '**results**' folder (label_atlas.nii).  
If registration is inaccurate, choose 'variation' from 1 to 4 and repeat cell.
